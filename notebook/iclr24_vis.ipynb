```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
df = pd.DataFrame({"M": [200,200,200,200,400,400,400,400,600,600,600,600,800,800,800,800], 
                   "m": [150,300,450,600,150,300,450,600,150,300,450,600,150,300,450,600],
                  "Accuracy": [68.32, 71.42, 73.84, 75.58, 
                                71.41, 75.60, 77.57, 78.66,
                                73.85, 77.59, 79.04, 79.76,
                                75.56, 78.68, 79.78, 80.26]}
                )
df['m'] = df['m'].astype('category')
# Create a line plot with markers
sns.set(style="white")
plt.figure(figsize=(10,10))  # Set the size of the plot
ax = sns.lineplot(data=df, x='M', y='Accuracy', hue='m', style='m', markers='^', dashes=False, linewidth = 2.5, markersize = 18)

# Add rug plot
sns.rugplot(data=df, x='M', ax=ax, color='black', height = 0.01)

# Set plot properties
ax.set_xlabel("Number of tasks (M)", fontsize=30)
ax.set_ylabel("Accuracy (%)", fontsize=30)
# Set x and y axis ticks
ax.set_xticks(range(200, 1000, 200)) # From 0 to 1000 with a step of 100
ax.set_yticks(range(70, 85, 5)) # From 70 to 90 with a step of 5
ax.tick_params(labelsize=30)
plt.legend(title='Samples per task (m)', title_fontsize='20', fontsize='18')


# Show the plot
plt.show()

fig = ax.get_figure()
fig.savefig("acc_vs_numTask.pdf", bbox_inches='tight')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
df = pd.DataFrame({"M": [200,200,200,200,400,400,400,400,600,600,600,600,800,800,800,800], 
                   "m": [150,300,450,600,150,300,450,600,150,300,450,600,150,300,450,600],
                  "Accuracy": [68.32, 71.42, 73.84, 75.58, 
                                71.41, 75.60, 77.57, 78.66,
                                73.85, 77.59, 79.04, 79.76,
                                75.56, 78.68, 79.78, 80.26]}
                )
df['M'] = df['M'].astype('category')
# Create a line plot with markers
sns.set(style="white")
plt.figure(figsize=(10,10))  # Set the size of the plot
ax = sns.lineplot(data=df, x='m', y='Accuracy', hue='M', style='M', markers='^', dashes=False, linewidth = 2.5, markersize = 18)

# Add rug plot
sns.rugplot(data=df, x='m', ax=ax, color='black', height = 0.01)

# Set plot properties
ax.set_xlabel("Samples per task (m)", fontsize=30)
ax.set_ylabel("Accuracy (%)", fontsize=30)
# Set x and y axis ticks
ax.set_xticks(range(150, 700, 150)) # From 0 to 1000 with a step of 100
ax.set_yticks(range(70, 85, 5)) # From 70 to 90 with a step of 5
ax.tick_params(labelsize=30)
plt.legend(title='Number of tasks (M)', title_fontsize='20', fontsize='18')


# Show the plot
plt.show()

fig = ax.get_figure()
fig.savefig("acc_vs_sampleTask.pdf", bbox_inches='tight')
```
